```python
import random

class Account:
    bank_menu = ['은행이름', '예 금 주', '계좌번호', '남은잔액', '입금횟수', '출금횟수']
    deposit_count = 1
    withdraw_count = 0

    def __init__(self, name, account_balance):
      self.client = ['SC은행']
      self.deposit_history_index = []
      self.withdraw_history_index = []
      self.client.append(name)
      self.account_number()
      self.client.append(account_balance)
      self.client.append(self.deposit_count)
      self.client.append(self.withdraw_count)
      self.deposit_history_index.append(account_balance)

    def format_currency(self, amount):
      format = '{:,.0f}'.format(amount)
      return format

    def display_info(self):
      print()
      for i in range(len(self.bank_menu)):
        if i == 3:
          print(i+1, Account.bank_menu[i], ' : ', self.format_currency(self.client[i]))
        else:
          print(i+1, Account.bank_menu[i], ' : ', self.client[i])

    def display_info_search(self, money):
      if self.client[3] >= money:
        self.display_info()

    def account_number(self):
      number = ''.join(str(random.randint(0, 9)) for x in range(12))
      self.account_number = number[:3] + '-' + number[3:6] + '-' + number[6:]
      self.client.append(self.account_number)

    def get_account_num(self, count):
      print("생성된 계좌 객체 개수 : ", count, "입니다")

    def deposit(self, money):
      print()
      print("예금주 :", self.client[1])
      print("입금 금액 :", money, '원')
      if money < 1:
        print("입금은 최소 1원 이상만 가능합니다.")
      elif money > 1:
        self.client[3] += money
        self.client[4] += 1
        self.deposit_history_index.append(money)

      if self.client[4] == 5:
        print("\n 축하합니다. 이자 1% 지급되었습니다.\n")
        self.client[3] += self.client[3] * 0.01

    def withdraw(self, money):
      print()
      print("예금주 :", self.client[1])
      print("출금 금액 :", money, '원')
      if self.client[3] < money:
        print("잔액이 부족합니다.")
      else:
        self.client[3] -= money
        self.client[5] += 1
        self.withdraw_history_index.append(money)

    def deposit_history(self):
      print(f"{self.client[1]}, 님의 입금내역 리스트")
      for i in range(len(self.deposit_history_index)):
        print(i, ":", self.deposit_history_index[i])

    def withdraw_history(self):
      print(f"{self.client[1]}, 님의 출금내역 리스트")
      for i in range(len(self.withdraw_history_index)):
        print(i, ":", self.withdraw_history_index[i])
```
